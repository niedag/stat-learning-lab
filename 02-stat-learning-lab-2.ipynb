```python
# For Loops

total = 0
for value in [3,2,19]:
    total += value
print('total is {0}'.format(total)) 
```

```python
total = 0
for value in [3,2,19]:
    for weight in [3,2,1]:
        total += value * weight
print('total is: {0}'.format(total))
```

```python
total = 0
for value, weight in zip([2,3,19],
                         [0.2,0.3,0.5]):
    total += weight * value
print('weighted avg is: {0}'.format(total))
```

```python
# String formatting
# Many data cleaning tasks involve manipulating and programmatically producing strings!
# Ex: we want to loop over the columns of a data frame and print the percent missing in each column

rng = np.random.default_rng(1)
A = rng.standard_normal((127,5))
M = rng.choice([0, np.nan], p = [0.8,0.2], size = A.shape)
A += M
D = pd.DataFrame(A, columns = ['food',
                               'bar',
                               'pickle',
                               'snack',
                               'popcorn'])
D[:3]
```

```python
for col in D.columns:
    template = 'Column "{0}" has {1:.2%} missing values'
    print(template.format(col, np.isnan(D[col]).mean()))
```
